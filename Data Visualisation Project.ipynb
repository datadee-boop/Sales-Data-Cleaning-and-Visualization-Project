```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns


df = pd.read_csv('/Users/danon/Downloads/Sales_Dataset.csv')
df
```

```python
1. 

# Sales Distribution by Product Category 🛍️

df.groupby(['Product_Category'])['Total_Amount'].sum().sort_values(ascending=False).plot(kind='bar' , figsize= (10,6), width=0.7, color='teal', edgecolor= 'black')

plt.title('Sales Distribution by Product Category')
plt.xlabel('Product Category')
plt.ylabel('Total Revenue')
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))

plt.xticks(rotation= 0)
plt.grid(axis= 'y', linestyle='--')
plt.tight_layout()
plt.show()
```

```python
2. 

# Sales Trend Over Time 📈

df['Order_Date'] = pd.to_datetime(df['Order_Date'])


df.groupby(['Order_Date'])['Total_Amount'].sum().plot(kind='line', figsize= (10,6), color= 'dodgerblue', marker='o')
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))

plt.title('Sales Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Total Sales Amount')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
3. 
# Sales by Payment Method 💳

df.groupby('Payment_Method')['Total_Amount'].sum().plot(kind='pie', figsize= (8,8), autopct = '%1.1f%%', startangle=20, cmap='cool')
plt.title('Sales Distribution by Payment Method')
plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
4.

# Revenue by Customer Region 🌍


df.groupby('Customer_Region')['Total_Amount'].sum().sort_values(ascending=True).plot(kind='barh', cmap='cool', figsize=(10,6), edgecolor='black')
plt.gca().get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))

plt.title('Revenue by Customer Region')
plt.ylabel('Customer Region')
plt.xlabel('Total Revenue')
plt.grid(axis= 'x', linestyle = '--')
plt.tight_layout()
plt.show()
```

```python
5. 

#Sales vs. Quantity Sold 📊


df.plot(x='Quantity', y='Total_Amount', cmap='spring', figsize= (10,6))
plt.title('Sales vs Quantity Sold')
plt.xlabel('Quantity Sold')
plt.ylabel('Total Sales Amount')
plt.grid(True)
```

```python
6 
# Customer Spending Distribution 💸

df.plot(y= 'Total_Amount', kind='hist', bins=20, edgecolor= 'black', color= 'crimson', figsize=(10,6))
plt.gca().get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))
plt.title('Distribution of Total Sales Amount')
plt.xlabel('Total Sales Amount')
plt.ylabel('Frequency')
plt.tight_layout()
plt.grid(True)
plt.show()
```

```python
7. 

#Top Products by Sales 📦

df.groupby('Product_Name')['Total_Amount'].sum().sort_values(ascending=True).plot(kind='barh', figsize=(10,6), color= 'mediumturquoise', edgecolor= 'limegreen')
plt.gca().get_xaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))
plt.grid(True)
plt.title('Top Products by Sales')
plt.xlabel('Total Sales Amount')
plt.ylabel('Product Name')
plt.tight_layout()
plt.show()
```

```python
8. 

# Heatmap of Sales by Product Category and Payment Method 🔥

heatmap_data = df.pivot_table(values='Total_Amount', index='Product_Category', columns='Payment_Method', aggfunc='sum')
plt.figure(figsize=(10,6))
sns.heatmap(heatmap_data, annot=True, fmt=".0f", cmap="cool", linewidths=0.5)

plt.title('Total Sales Amount by Product Category and Payment Method')
plt.xlabel('Payment Method')
plt.ylabel('Product Category')
plt.tight_layout()
plt.xticks(rotation=0)

plt.show()
```

```python
9. 

# Boxplot of Sales Distribution by Region 📦🌍

plt.figure(figsize=(10, 6))
plt.gca().get_yaxis().set_major_formatter(plt.FuncFormatter(lambda x, loc: "${:,}".format(int(x))))

sns.boxplot(x='Customer_Region', y='Total_Amount', data=df, color= 'turquoise')

plt.title('Sales Distribution by Customer Region')
plt.xlabel('Customer Region')
plt.ylabel('Total Sales Amount')
plt.show()

```

```python
10. 

#Total Sales by Day of the Week 🗓️


df['Day_of_Week'] = df['Order_Date'].dt.day_name()

weekly_sales = df.groupby('Day_of_Week')['Total_Amount'].sum().reset_index()

# Sort the days in the correct order
ordered_days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
weekly_sales['Day_of_Week'] = pd.Categorical(weekly_sales['Day_of_Week'], categories=ordered_days, ordered=True)
weekly_sales = weekly_sales.sort_values('Day_of_Week')

plt.figure(figsize=(10, 6))
sns.barplot(x='Day_of_Week', y='Total_Amount', data=weekly_sales, palette='magma', hue= 'Day_of_Week', edgecolor='black', legend=False)
plt.title('Total Sales by Day of the Week')
plt.xlabel('Day of the Week')
plt.ylabel('Total Sales Amount')
plt.xticks(rotation=0)
plt.grid(True)
plt.tight_layout()
plt.show()
```
